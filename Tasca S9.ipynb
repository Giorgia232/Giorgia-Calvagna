**Nivel 1 - Ejercicio 1: Calculadora del índice de masa corporal (IMC)**

```python
def calcular_imc():
    # TRY: intentamos hacer esto
    try:
        # Pedimos datos al usuario y los convertimos a números decimales (float)
        peso = float(input("Introduce el peso (kg): "))
        altura = float(input("Introduce la altura (m): "))
        
        # Validamos que sean valores razonables (positivos)
        if peso <= 0 or altura <= 0:
            print("El peso y la altura deben ser positivos.")
            return  # Salimos de la función
        
        # Calculamos el IMC
        imc = peso / (altura ** 2)  # **2 = al cuadrado
        
        # Clasificamos según el valor del IMC
        if imc < 18.5:
            categoria = "Bajo peso"
        elif imc < 25:
            categoria = "Peso normal"
        elif imc < 30:
            categoria = "Sobrepeso"
        else:
            categoria = "Obesidad"
        
        # Mostramos resultado con 2 decimales (.2f)
        print(f"IMC: {imc:.2f} - {categoria}")
    
    
    # EXCEPT: si hay error al convertir a número, mostramos aviso
    except ValueError:
        print("Introduce números válidos.")

# Llamamos a la función para que se ejecute
calcular_imc()
```

**Nivel 1 - Ejercicio 2: Convertidor de temperaduras**

```python
def convertir_temperatura(valor, unidad_origen, unidades_destino=["F", "K"]):

    # Validamos que el valor sea numérico
    try:
        temp = float(valor)
    except ValueError:
        print("Advertencia: El valor introducido no es un numero valido.")
        return {}

    # Normalizamos la unidad: tomamos la primera letra en mayúscula ('celsius' -> 'C')
    unidad_origen = unidad_origen.upper()
    if unidad_origen not in {"C", "F", "K"}:
        print(f"Advertencia: Unidad '{unidad_origen}' no valida. Usa C, F o K.")
        return {}

    # Convertimos primero a Celsius como unidad base común
    if unidad_origen == "C":
        celsius = temp
    elif unidad_origen == "F":
        celsius = (temp - 32) * 5/9
    elif unidad_origen == "K":
        celsius = temp - 273.15

    # Desde Celsius calculamos todas las conversiones posibles
    conversiones = {
        "C": celsius,
        "F": celsius * 9/5 + 32,
        "K": celsius + 273.15
    }

    # Guardamos solo las conversiones pedidas en un diccionario, excluyendo la unidad de origen
    resultados = {}
    for unidad in unidades_destino:
        unidad = unidad.upper()
        if unidad in conversiones and unidad != unidad_origen:
            resultados[unidad] = round(conversiones[unidad], 2)

    return resultadosD


# Ejecucion
resultado = convertir_temperatura(100, "C")
print(resultado)
```

**Nivel 1 - Ejercicio 3: Contador de palabras de un texto**

```python
def contador_palabras(texto):
    # Validación: Comprobamos que sea un texto (string)
    if not isinstance(texto, str) or len(texto) == 0:
        return "ERROR: Por favor, introduce un texto válido."

    # Limpieza: quitamos signos de puntuación básicos y pasamos el texto a minúsculas
    # Esto evita que "Hola" y "hola," se cuenten como palabras diferentes
    caracteres_a_quitar = ",.?!¿:;¡"
    texto_limpio = texto.lower()
    for caracter in caracteres_a_quitar:
        texto_limpio = texto_limpio.replace(caracter, "")

    # Segmentación: dividimos el texto en una lista de palabras
    palabras = texto_limpio.split()
    
    # Conteo: usamos un diccionario para contar las frecuencias
    frecuencias = {}
    for palabra in palabras:
        frecuencias[palabra] = frecuencias.get(palabra, 0) + 1
    
    # Sumamos el largo de todas las palabras y dividimos por el total de palabras para obtener la longitud media
    total_letras = sum(len(palabra) for palabra in palabras)
    media = total_letras / len(palabras) if palabras else 0

    return frecuencias, round(media, 2)

# Ejecución
texto_usuario = "Hola, ¿cómo va? Hola de nuevo, te deseo un feliz día!"
conteo, longitud_media = contador_palabras(texto_usuario)

print(f"Frecuencia de palabras: {conteo}")
print(f"Longitud media de las palabras: {longitud_media}")
```

**Nivel 1 - Ejercicio 4: Diccionario inverso**

```python
def diccionario_inverso(diccionario):

    # Creamos el diccionario donde guardaremos el resultado invertido
    resultado = {}
    hay_duplicados = False

    # Recorremos el diccionario original par a par (clave, valor)
    for clave, valor in diccionario.items():

        # Si el valor no existe aun en el resultado, lo añadimos normalmente
        if valor not in resultado:
            resultado[valor] = clave

        # Si el valor ya existe, es un duplicado
        else:
            hay_duplicados = True

            # Si lo que hay guardado no es una lista, lo convertimos en lista
            if not isinstance(resultado[valor], list):
                resultado[valor] = [resultado[valor]]

            # Añadimos la nueva clave a la lista
            resultado[valor].append(clave)

    # Si hubo duplicados, mostramos un aviso con los valores afectados
    if hay_duplicados:
        print("ADVERTENCIA: Se han detectado valores duplicados.")
        for valor, claves in resultado.items():
            if isinstance(claves, list):
                print(f"  El valor '{valor}' aparece en las claves: {claves}")

    return resultado


# Ejecución caso 1: sin duplicados
print(diccionario_inverso({'a': 1, 'b': 2, 'c': 3}))

# Ejecución caso 2: con duplicados
print(diccionario_inverso({'x': 'apple', 'y': 'banana', 'z': 'banana'}))
```

**Nivel 2 - Ejercicio 1: Contador y organizador de palabras de un texto**

```python
def contador_organizado(archivo):
    try:
        # Abrimos y leemos el archivo
        with open(archivo, 'r', encoding='utf-8') as archivo_abierto:
            texto = archivo_abierto.read().lower()

        # Eliminamos los signos de puntuacion mas comunes
        for signo in ",.?!:;()¿¡\"":
            texto = texto.replace(signo, "")

        # Dividimos el texto en palabras y las ordenamos alfabeticamente
        palabras = sorted(texto.split())

        # Creamos el diccionario anidado: letra inicial -> {palabra: frecuencia}
        resultado = {}
        for palabra in palabras:
            letra_inicial = palabra[0]
            if letra_inicial not in resultado:
                resultado[letra_inicial] = {}
            resultado[letra_inicial][palabra] = resultado[letra_inicial].get(palabra, 0) + 1

        return resultado

    except FileNotFoundError:
        return f"Error: No se encontró el archivo '{archivo}'."


# Ejecución
resultado = contador_organizado("tu me quieres blanca.txt")

for letra_inicial, grupo_palabras in resultado.items():
    print(f"\n[{letra_inicial.upper()}]")
    for palabra, frecuencia in grupo_palabras.items():
        print(f"  {palabra}: {frecuencia}")
```

**Nivel 2 - Ejercicio 2: Conversión de tipo de datos**

```python
def separar_elementos(lista_de_datos):

    # Creamos las dos listas de resultados
    elementos_convertidos = []
    elementos_no_convertibles = []

    # Recorremos cada elemento de la lista
    for elemento in lista_de_datos:

        # Si el elemento es una lista o tupla, recorremos su contenido
        if type(elemento) in [list, tuple]:
            for sub_elemento in elemento:
                try:
                    elementos_convertidos.append(float(sub_elemento))
                except (ValueError, TypeError):
                    elementos_no_convertibles.append(sub_elemento)

        # Si no, intentamos convertirlo directamente
        else:
            try:
                elementos_convertidos.append(float(elemento))
            except (ValueError, TypeError):
                elementos_no_convertibles.append(elemento)

    return elementos_convertidos, elementos_no_convertibles


# Ejecución
datos = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]
convertidos, no_convertibles = separar_elementos(datos)

print(f"Elementos convertidos a float: {convertidos}")
print(f"Elementos no convertibles: {no_convertibles}")
```

```python
import numpy as np
import string
import pyperclip

def crear_contrasena(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):
    # Creamos un "almacén" de caracteres disponibles
    caracteres_disponibles = ""
    
    if mayusculas:
        caracteres_disponibles += string.ascii_uppercase  # A-Z
    if minusculas:
        caracteres_disponibles += string.ascii_lowercase  # a-z
    if numeros:
        caracteres_disponibles += string.digits           # 0-9
    if signos:
        caracteres_disponibles += string.punctuation      # !@#$%...

    # Validación de seguridad: si no se elige ningún tipo, lanzamos un aviso
    if not caracteres_disponibles:
        return "Error: No se puede generar la contraseña porque no se ha seleccionado ningún tipo de carácter (mayúsculas, minúsculas, números o signos)."

    # Conversión para NumPy
    # Convertimos el almacén en una lista de caracteres individuales
    lista_de_caracteres = list(caracteres_disponibles)

    # Generación aleatoria profesional con NumPy
    # 'choice' elige elementos al azar de la lista según la longitud deseada
    contrasena_aleatoria = np.random.choice(lista_de_caracteres, size=longitud)

    # Unimos la lista en un solo texto (string)
    contrasena_final = "".join(contrasena_aleatoria)

    # Copiar automáticamente al portapapeles
    pyperclip.copy(contrasena_final)
    
    return contrasena_final

# --- Bloque de Ejecución ---
# Probamos la función con 10 caracteres y los parámetros elegidos
mi_nueva_contrasena = crear_contrasena(10, True, True, True, False)

print(f"Tu nueva contraseña es: {mi_nueva_contrasena}")
print("¡Nota: La contraseña ya ha sido copiada a tu portapapeles!")
```

**Nivel 3 - Ejercicio 1: Proceso de datos simple**

```python
def obtener_puntos(item):
    # Esta funcion auxiliar le dice a sorted() que ordene por el segundo elemento (los puntos)
    return item[1]


def analizar_partidos(nombre_archivo):
    goles_hechos = {}
    goles_recibidos = {}
    puntos = {}

    try:
        with open(nombre_archivo, 'r', encoding='utf-8') as archivo:
            for linea in archivo:
                linea = linea.strip()
                if not linea:
                    continue

                # Separamos por tabulador: ['Manlleu', '0-1', 'Granollers']
                datos = linea.split('\t')
                equipo_local = datos[0]
                marcador = datos[1].split('-')
                goles_local = int(marcador[0])
                goles_visitante = int(marcador[1])
                equipo_visitante = datos[2]

                # Inicializamos los equipos si aparecen por primera vez
                for equipo in [equipo_local, equipo_visitante]:
                    if equipo not in puntos:
                        puntos[equipo] = 0
                        goles_hechos[equipo] = 0
                        goles_recibidos[equipo] = 0

                # Actualizamos los goles
                goles_hechos[equipo_local] += goles_local
                goles_hechos[equipo_visitante] += goles_visitante
                goles_recibidos[equipo_local] += goles_visitante
                goles_recibidos[equipo_visitante] += goles_local

                # Asignamos puntos según el resultado
                if goles_local > goles_visitante:
                    puntos[equipo_local] += 3
                elif goles_local < goles_visitante:
                    puntos[equipo_visitante] += 3
                else:
                    puntos[equipo_local] += 1
                    puntos[equipo_visitante] += 1

    except FileNotFoundError:
        print(f"Error: No se encontro el archivo '{nombre_archivo}'.")
        return

    # Calculamos los resultados finales
    equipo_mas_goleador = max(goles_hechos, key=goles_hechos.get)
    equipo_mas_goleado = max(goles_recibidos, key=goles_recibidos.get)
    clasificacion = sorted(puntos.items(), key=obtener_puntos, reverse=True)

    return goles_hechos, equipo_mas_goleador, equipo_mas_goleado, clasificacion


# Ejecucion
goles, mas_goleador, mas_goleado, clasificacion = analizar_partidos("historic partits.txt")

print(f"Maximo goleador: {mas_goleador} ({goles[mas_goleador]} goles)")
print(f"Maximo goleado: {mas_goleado}")
print("\n--- CLASIFICACION ---")
for equipo, pts in clasificacion:
    print(f"  {equipo}: {pts} pts")
```
